```python
import pandas as pd
```

```python
data_frame = pd.read_csv('xbtusd_4H_last_2021_02_10.csv')
```

```python
data_frame.head()
```

```python
df = data_frame.set_index(['T'])
df.head()
```

```python
df.index
```

```python
df.index = df.index.astype('datetime64[ns]')
```

```python
df.index
```

```python
df = df.resample('4H').agg({
    'O': 'first',
    'H': 'max',
    'L': 'min',
    'C': 'last',
    'V': 'sum'
})
df
```

```python
df = df.reset_index()
df
```
