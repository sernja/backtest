```python
import pandas as pd
```

```python
df = pd.read_csv('../XBTUSD-1h-data.csv')
```

```python
df
```

### Remove 'symbol', 'trades', 'vwap', 'lastSize', 'turnover', 'homeNotional', 'foreignNotional'

```python
df = df.drop(columns=['symbol', 'trades', 'vwap', 'lastSize', 'turnover', 'homeNotional', 'foreignNotional'])
df
```

### Convert timestamp to utf-8

```python
df['timestamp'] = df['timestamp'].str.decode("utf-8")
df['timestamp']
```

### Remove time after '+'

```python
df['timestamp'] = df['timestamp'].str.split('+').str[0]
df['timestamp']
```

### Rename all columns

```python
df = df.rename(columns={"timestamp": "T", "open": "O", "high": "H", "low": "L", "close":"C", "volume": "V"})
```
